```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Penguins = pd.read_csv('/content/penguins_lter.csv')
Penguins
```

```python
#general info of dataset
Penguins.info()
```

```python
#unique values present
Penguins.nunique()
```

```python
#checking for duplicates
Penguins.duplicated().sum()
```

```python
#checking for null values
Penguins.isnull().sum()
```

```python
#dropping columns
Penguins.drop(columns=['Comments','Delta 15 N (o/oo)','Delta 13 C (o/oo)','studyName','Sample Number','Stage','Individual ID','Region'], inplace=True)
```

```python
#getting mean value and filling null values with mean
```

```python
a=Penguins['Culmen Length (mm)'].mean()
a
```

```python
Penguins['Culmen Length (mm)'].fillna(a,inplace=True)
```

```python
b = Penguins['Culmen Depth (mm)'].mean()
b
```

```python
Penguins['Culmen Depth (mm)'].fillna(b,inplace=True)
```

```python
c=Penguins['Flipper Length (mm)'].mean()
c
```

```python
Penguins['Flipper Length (mm)'].fillna(c,inplace=True)
```

```python
d=Penguins['Body Mass (g)'].mean()
d
```

```python
Penguins['Body Mass (g)'].fillna(d,inplace=True)
```

```python
#finding mode sex
mode_sex = Penguins['Sex'].mode()
mode_sex
```

```python
#filling null with mode
mode_sex = Penguins['Sex'].mode()[0]
Penguins['Sex'].fillna(mode_sex, inplace=True)
```

```python
#checking for null values again
Penguins.isnull().sum()
```

```python
# convert the 'Date' column to datetime format
Penguins['Date Egg']= pd.to_datetime(Penguins['Date Egg'])

# Check the format of 'Date' column
Penguins.info()
```

```python
#converting Body Mass from (g) to (kg)
Penguins['Body Mass(kg)'] = Penguins['Body Mass (g)'] / 1000
```

```python
#dropping Body Mass(g) column
Penguins.drop(columns = ['Body Mass (g)'])
```

```python
#capitalize column headers
Penguins.rename(columns=lambda x: x.capitalize(), inplace=True)
```

```python
#split scientific name from generic name
Penguins['Species']=Penguins['Species'].str.split("(").str[0]
```

```python
Penguins.tail(3)
```

```python
#dropping a value

Penguins= Penguins.drop(Penguins[Penguins['Sex'] == '.'].index)
```

```python
#importing libraaries
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#value count of penguin species
Penguins['Species'].value_counts()
```

```python
#grouping by species to get population
Penguins['Population'] = Penguins.groupby('Species')['Species'].transform('count')

#plot for species distribution
sns.barplot(x='Species',y='Population',hue='Species',data=Penguins)
plt.title('Penguin Species Distribution')
plt.show()
```

```python
#plot for species distribution per island
sns.barplot(x='Island',y='Population',hue='Species',data=Penguins)
plt.title('Penguin Species Distribution per Island')
plt.show()
```

```python
#plotting body mass by sex
plt.figure(figsize=(8, 6))
sns.boxplot(x='Sex', y='Body mass(kg)', data=Penguins, hue='Sex')
plt.title('Body Mass by Sex')
plt.xlabel('Sex')
plt.ylabel('Body Mass (kg)')
plt.show()
```

```python
#plot body mass by sex per species
sns.boxplot(x='Sex',y='Body mass(kg)',hue='Species',data=Penguins)
plt.title('Body Mass by Sex for Each Species')
plt.show()
```

```python
#plotting clutch completion by sex
plt.figure(figsize=(10, 6))


sns.countplot(x='Clutch completion', hue='Sex', data=Penguins, palette=["lightblue", "pink"],
              order=Penguins['Clutch completion'].value_counts().index,
              hue_order=Penguins['Sex'].value_counts().index,
              dodge=True)

plt.title('Clutch Completion by Sex')
plt.xlabel('Clutch Completion')
plt.ylabel('Count')
plt.show()
```

```python
#plot for clutch completion per island
plt.figure(figsize=(12, 6))
sns.countplot(x='Island', hue='Clutch completion', data=Penguins)
plt.title('Clutch Completion Distribution on Different Islands')
plt.xlabel('Island')
plt.ylabel('Count')
plt.show()
```

```python
#correlation between culmen length and body mass per species
sns.scatterplot(x='Body mass(kg)', y='Culmen length (mm)',hue='Species', data=Penguins)
plt.title('The Relationship Between Culmen Length and Body Mass by Penguin Species')
plt.show()
```

```python
#correlation between culmen depth and body mass per species
sns.scatterplot (x='Body mass(kg)',y='Culmen depth (mm)',hue='Species',data=Penguins)
plt.title('The Relationship Between Culmen Depth and Body Mass by Penguin Species')
plt.show()
```

```python
#correlation between flipper length and body mass per species
sns.scatterplot(x='Body mass(kg)', y='Flipper length (mm)',hue='Species', data=Penguins)
plt.title('The Relationship Between Flipper Length and Body Mass by Penguin Species')
plt.show()
```
